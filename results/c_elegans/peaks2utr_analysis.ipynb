```python
import os
from pathlib import Path

import gffutils
import pandas as pd
import sqlite3

in_dir = "input"
out_dir = "output"
Path(out_dir).mkdir(exist_ok=True)
```

## Create GFF databases and read in features

```python
def create_db(path, db_filename, **kwargs):
    """
    Create/connect db for given gff file path
    """
    if not os.path.isfile(os.path.join(out_dir, db_filename)):
        db = gffutils.create_db(path, os.path.join(out_dir, db_filename), force=True, **kwargs)
    else:
        db = sqlite3.connect(os.path.join(out_dir, db_filename), check_same_thread=False)
        db = gffutils.FeatureDB(db)
    return db

# WormBase release WS283 canonical three_prime_UTR for C. elegans chromosome I
control_fn = os.path.join(in_dir, "WormBase.I.three_prime_UTR.gff3")
db_control = create_db(control_fn, "control.db")
control = list(db_control.all_features())

# Output from peaks2utr run
test_fn = os.path.join(in_dir, "peaks2utr_output.gff3")
db_test = create_db(test_fn, "test.db")
test = list(db_test.all_features())

# WormBase release WS283 canonical mRNA transcripts for C. elegans chromosome I
genes_fn = os.path.join(in_dir, "WormBase.I.mRNA.gff3")
db_genes = create_db(genes_fn, "genes.db")
genes = list(db_genes.all_features())
```

## Prepare test dataset

Create a dataframe of peaks2utr test UTRs with a "gene id" index.

```python
tdf = pd.DataFrame([vars(t) for t in test], columns=["start", "end", "strand", "attributes"])
tdf['name'] = tdf["attributes"].apply(lambda x: x.get("sequence_name")[0])
tdf["gene"] = tdf["attributes"].apply(lambda x: x.get("Name")[0])
test_utrs = tdf.drop(["name", "attributes"], axis=1).set_index("gene")
test_utrs
```

## Prepare control dataset

C. elegans contains alternative mRNA transcripts for each gene, and the control UTRs refer only to these mRNA transcript "names" in their attributes. In this instance, peaks2utr has been configured to call UTRs on the parent gene feature itself, rather than the child transcripts. Therefore, it will be helpful to create a sequence name -> gene id mapping dataframe to allow comparisions between the two datasets.

```python
gdf = pd.DataFrame([vars(c) for c in genes], columns=["featuretype", "start", "end", "strand", "attributes"])
gdf = gdf[gdf["featuretype"] == "mRNA"]
gdf["gene"] = gdf["attributes"].apply(lambda x: x.get("Parent")[0].split(":")[-1])
gdf["name"] = gdf["attributes"].apply(lambda x: x.get("Name")[0])
gdf = gdf[["name", "gene"]]
gdf
```

Create a dataframe of canonical WormBase control UTRs with a "name" column consisting of the parent gene's name.

```python
cdf = pd.DataFrame([vars(c) for c in control], columns=["start", "end", "source", "strand", "attributes"])
cdf = cdf[cdf.source == "WormBase"]
cdf['name'] = cdf["attributes"].apply(lambda x: x.get("Parent")[0].split(":")[-1])
cdf = cdf.drop(['attributes', "source"], axis=1)
cdf = pd.merge(cdf, gdf, on="name")
cdf
```

As peaks2utr has used the --override-utr flag, we know that the start/end base will match the start/end base of the 3' UTR for the innermost mRNA transcript (on the forward/reverse strand, respectively). Therefore, we can aggregate each gene by taking the minimum start base and maximum end base to capture the full extent of the UTR for that gene.

```python
control_utrs = cdf.groupby(["gene", "strand"]).agg({"start": min, "end": max}).reset_index("strand")
control_utrs
```

# Compare the two datasets

Now that we have two comparable datasets with paired UTRs, we can calculate some statistical differences. We determine length of each UTR in both dataframes, and determine differences of test UTR length to true UTR length.

```python
control_utrs["len"] = abs(control_utrs["end"] - control_utrs["start"])
test_utrs["len"] = abs(test_utrs["end"] - test_utrs["start"])
test_utrs["difference"] = (test_utrs["len"] - control_utrs["len"])
test_utrs
```

## Supplementary Table 2 statistics

```python
total_test = len(test_utrs)
total_control = len(control_utrs)

genes_missed_utr = control_utrs.index.difference(test_utrs.index)
num_missed_utrs = len(genes_missed_utr)
genes_no_utr = set(gdf["gene"]).difference(control_utrs.index)
novel_utr_genes = genes_no_utr.intersection(test_utrs.index)
num_novel_utrs = len(novel_utr_genes)

BP1 = 50
BP2 = 200
genes_matched_utr =  test_utrs[abs(test_utrs["difference"]) <= BP1].index
num_matched = len(genes_matched_utr)
num_reduced = len(test_utrs[test_utrs['difference'] < -BP1])
num_extended1 = len(test_utrs[(test_utrs['difference'] > BP1) & (test_utrs['difference'] <= BP2)])
num_extended2 = len(test_utrs[test_utrs['difference'] > BP2])

print("Canonical utrs that are missed by peaks2utr: {} ({:.2f}%)".format(num_missed_utrs, 100*num_missed_utrs/total_control))
print("peaks2utr utrs that match true utr to within {} bases: {} ({:.2f}%)".format(BP1, num_matched, 100*num_matched/total_control))
print("peaks2utr utrs that reduce true utr by >{} bases: {} ({:.2f}%)".format(BP1, num_reduced, 100*num_reduced/total_control))
print("peaks2utr utrs that extend true utr by between {} and {} bases: {} ({:.2f}%)".format(BP1, BP2, num_extended1, 100*num_extended1/total_control))
print("peaks2utr utrs that extend true utr by >{} bases: {} ({:.2f}%)".format(BP2, num_extended2, 100*num_extended2/total_control))
print("peaks2utr utrs called where no true utr existed previously: {}".format(num_novel_utrs))
print("Total peaks2utr UTRs: {}".format(total_test))
```

Save lists of gene ids with novel, matched and missed 3' UTRs called by peaks2utr

```python
with open(os.path.join(out_dir, "novel_utr_genes_peaks2utr.txt"), "w") as f:
    for g in sorted(list(novel_utr_genes)):
        f.write(g + "\n")

with open(os.path.join(out_dir, "matched_utr_genes_peaks2utr.txt"), "w") as f:
    for g in sorted(list(genes_matched_utr)):
        f.write(g + "\n")

with open(os.path.join(out_dir, "missed_utr_genes_peaks2utr.txt"), "w") as f:
    for g in sorted(list(genes_missed_utr)):
        f.write(g + "\n")
```
