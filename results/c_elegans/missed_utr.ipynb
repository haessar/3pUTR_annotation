```python
import os

import matplotlib.pyplot as plt
from matplotlib_venn import venn3_circles, venn3

out_dir = "output"
```

```python
missed_peaks2utr_file = os.path.join(out_dir, "missed_utr_genes_peaks2utr.txt")
missed_getutr_file = os.path.join(out_dir, "missed_utr_genes_getutr.txt")
missed_utrme_file = os.path.join(out_dir, "missed_utr_genes_utrme.txt")

if not all([
    os.path.exists(missed_peaks2utr_file),
    os.path.exists(missed_getutr_file),
    os.path.exists(missed_utrme_file)
]):
    raise FileNotFoundError("Please run all three analysis notebooks in full to produce the files required for this notebook")

def txt_to_list(file):
    with open(file, "r") as f:
        return f.read().splitlines()

peaks2utr = set(txt_to_list(missed_peaks2utr_file))
getutr = set(txt_to_list(missed_getutr_file))
utrme = set(txt_to_list(missed_utrme_file))
```

Plot three-way Venn diagram to illustrate the extent to which each tool called a shared novel UTR

```python
venn3([peaks2utr, getutr, utrme], ["peaks2utr", "GETUTR", "UTRme"], set_colors=("blue", "yellow","fuchsia"))
venn3_circles([peaks2utr, getutr, utrme])
plt.suptitle("\"Missing\" 3' UTR intersections by each tool", fontsize=16)
plt.savefig(os.path.join(out_dir, "missed_utr_venn.png"))
```
