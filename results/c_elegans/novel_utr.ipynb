```python
import os

import matplotlib.pyplot as plt
from matplotlib_venn import venn3_unweighted, venn3_circles

out_dir = "output"
```

```python
novel_peaks2utr_file = os.path.join(out_dir, "novel_utr_genes_peaks2utr.txt")
novel_getutr_file = os.path.join(out_dir, "novel_utr_genes_getutr.txt")
novel_utrme_file = os.path.join(out_dir, "novel_utr_genes_utrme.txt")

if not all([
    os.path.exists(novel_peaks2utr_file),
    os.path.exists(novel_getutr_file),
    os.path.exists(novel_utrme_file)
]):
    raise FileNotFoundError("Please run all three analysis notebooks in full to produce the files required for this notebook")

def txt_to_list(file):
    with open(file, "r") as f:
        return f.read().splitlines()

peaks2utr = set(txt_to_list(novel_peaks2utr_file))
getutr = set(txt_to_list(novel_getutr_file))
utrme = set(txt_to_list(novel_utrme_file))
```

Plot three-way Venn diagram to illustrate the extent to which each tool called a shared novel UTR

```python
venn3_unweighted([peaks2utr, getutr, utrme], ["peaks2utr", "GETUTR", "UTRme"], set_colors=("blue", "yellow","fuchsia"))
venn3_circles(subsets=(1,)*7)
plt.suptitle("Novel 3' UTRs intersected by each tool", fontsize=16)
plt.savefig(os.path.join(out_dir, "novel_utr_venn.png"))
```

List of all genes with a novel UTR predicted by each tool

```python
sorted(utrme.intersection(peaks2utr, getutr))
```
